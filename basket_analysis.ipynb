```python
import pandas as pd
import numpy as np
import datetime
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
```

```python
df = pd.read_csv('C:\\Users\\grace\\Desktop\\0701_basket_02.csv')
```

```python
# 將date由int轉為datetime格式
j = 0
for i in df['dd']:
    #print(j,i)
    i = str(int(i))
    dd = datetime.datetime.strptime(i, '%Y%m%d').date()
    #print(dd)
    j = j+ 1
    #print(j,i)
```

```python
#pivot_table
basket = df[['user_id','name','amount']].pivot_table(index='user_id', columns='name', aggfunc='sum').fillna(0)
```

```python
#將購買多個的轉為1，如果A用戶買了5個B商品仍計為1
def encode_units(x):
    if x <= 0:
        return 0
    if x >= 1:
        return 1
```

```python
basket_sets = basket.applymap(encode_units)
```

```python
basket_sets
```

```python
#可調整 min_support=0.002
frequent_itemsets = apriori(basket_sets, min_support=0.002, use_colnames=True)
rules = association_rules(frequent_itemsets, metric='lift', min_threshold=1)
rules.sort_values('lift',ascending = False).head(50)
```
